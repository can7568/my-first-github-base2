```python
import numpy as np
import pandas as pd
# from tabulate import tabulate
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import json
```

```python
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
penguins = sns.load_dataset("penguins")
```

```python
sns.barplot(x='class', y='survived', data=titanic)
```

```python
sns.set_style('darkgrid') # 스타일 설정(dark, darkgrid, white,whithgrid,ticks)
```

```python
# 그래프 객체 생성(도화지 한장 만든 것임)
fig = plt.figure(figsize=(10,5))
# 도화지 분할(add_subplot(행, 열, 인덱스))
sect1 = fig.add_subplot(1,3,1)
sect2 = fig.add_subplot(1,3,2)
sect3 = fig.add_subplot(1,3,3)
# 부분 섹터별 설정
sect1.set_title("aaaa")
sect2.set_title("bbbb")
sect3.set_title("cccc")
# 섹터에 그래프 그리기
# hue:범레, 데이터분류
# dodge: 항목별 누적여부(True:개별, False:누적)
sns.barplot(x='sex', y='survived', data=titanic, ax=sect1)
sns.barplot(x='sex', y='survived', hue='class',
data=titanic, ax=sect2)
sns.barplot(x='sex', y='survived', hue='class',
dodge=False, data=titanic, ax=sect3)
```

```python
# 아래 내용을 위 그림처럼 barplot 3가지 유향으로 그리세요
# 데이터요소: x - who, y - survived
# 그래프별 제목: 임의로 부여하세요.
# sect1: x,y 값만
# sect2: "hue: class" 추가
# sect3: sect2를 누적그래프로
fig = plt.figure(figsize=(10,5))
# 도화지 분할(add_subplot(행, 열, 인덱스))
sect1 = fig.add_subplot(1,3,1)
sect2 = fig.add_subplot(1,3,2)
sect3 = fig.add_subplot(1,3,3)
# 부분 섹터별 설정
sect1.set_title("1")
sect2.set_title("2")
sect3.set_title("3")
# 섹터에 그래프 그리기
# hue:범레, 데이터분류
# dodge: 항목별 누적여부(True:개별, False:누적)
sns.barplot(x='who', y='survived',data=titanic, ax=sect1)
sns.barplot(x='who', y='survived', hue='class',data=titanic, ax=sect2)
sns.barplot(x='who', y='survived', hue='class',dodge=False, data=titanic, ax=sect3)
```
