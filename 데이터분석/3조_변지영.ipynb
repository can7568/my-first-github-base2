```python
# LAB
# 아래 내용을 위 그림처럼 barplot 3가지 유형으로 그리세요
# 데이터요소: x - who, y - survived
# 그래프별 제목: 임의로 부여하세요.
# sect1: x,y 값만
# sect2: "hue: class" 추가
# sect3: sect2를 누적그래프로

fig = plt.figure(figsize=(10,5))

sect1 = fig.add_subplot(1,3,1)
sect2 = fig.add_subplot(1,3,2)
sect3 = fig.add_subplot(1,3,3)

sect1.set_title("1")
sect2.set_title("2")
sect3.set_title("3")

sns.barplot(x='who', y='survived',data=titanic, ax=sect1)
sns.barplot(x='who', y='survived', hue='class',data=titanic, ax=sect2)
sns.barplot(x='who', y='survived', hue='class',dodge=False, data=titanic, ax=sect3)
```

```python
# LAB
# 아래 내용을 위 그림처럼 countplot 3가지 유향으로 그리세요
# 데이터요소: x - embarked
# 그래프별 제목: 임의로 부여하세요.
# sect1: x,y 값만
# sect2: "hue: class" 추가
# sect3: sect2를 누적그래프로
fig = plt.figure(figsize=(10,5))

sect1 = fig.add_subplot(1,3,1)
sect2 = fig.add_subplot(1,3,2)
sect3 = fig.add_subplot(1,3,3)

sect1.set_title("1")
sect2.set_title("2")
sect3.set_title("3")

sns.countplot(x='embarked',data=titanic, ax=sect1)
sns.countplot(x='embarked', hue='class',data=titanic, ax=sect2)
sns.countplot(x='embarked', hue='class',dodge=False, data=titanic, ax=sect3)
```

```python
# LAB
# 아래 내용을 위 그림처럼 stripplo,swarmplott 유형으로 그리세요
# 데이터요소: x embarked, y-age
# 그래프별 제목: 임의로 부여하세요.
# sect1: x,y 값만
# sect2: y축 'fare'값으로
# sect3: sect2를 swarmplot 으로 그리기
fig = plt.figure(figsize=(10,5))

sect1 = fig.add_subplot(1,3,1)
sect2 = fig.add_subplot(1,3,2)
sect3 = fig.add_subplot(1,3,3)

sect1.set_title("1")
sect2.set_title("2")
sect3.set_title("3")

sns.stripplot(x='embarked', y='age',data=titanic, ax=sect1)
sns.stripplot(x='embarked', y='fare',data=titanic, ax=sect2)
sns.swarmplot(x='embarked', y='fare', data=titanic, ax=sect3)
```
