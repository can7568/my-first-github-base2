```python
#12-3 외판원문제(TSP)_분기한정법
import sys
from queue import PriorityQueue
from copy import deepcopy

class Tree:
    def __init__(self,level):
        self.level = level #출발노드의 레벌 = 0
        self.path = []
        #self.weight = weight
        #self.profit = profit
        self.bound = 0
    
    def isIn(self, v): #path에 정점(v)이 있는지 비교
        for i in range(len(self.path)):
            if(self.path[i] == v):
                return True
        return False
    
    def __lt__(self, other): #우선순위 큐 오류
        return self.bound < other.bound

def length(path): #유튜브 참고
    total = 0
    prev = 1
    for i in range(len(path)):
        if (i != 0):
            prev = path[i - 1]
        total += W[prev][path[i]] 
        prev = path[i]
    return total

def hasOutgoing(v,path): # 0 ~ n-1
    for i in range(0,len(path)-1):
        if path[i] == v:
            return True
    return False

def hasIncoming(v,path): # 1 ~ n
    for i in range(1,len(path)):
        if path[i] == v:
            return True
    return False

def bound(v): #하한값
    lower = length(v.path)
    for i in range(1,n+1):
        if(hasOutgoing(i, v.path)):
            continue
        min = INF
        for j in range(1,n+1):
            if i ==j:
                continue
            if hasIncoming(j, v.path):
                continue
            if (j == 1 and i == v.path[len(v.path) - 1]):
                continue
            if min > W[i][j]:
                min = W[i][j]
        lower += min
    return lower

def travel (v):
    global minlength, opttour
    PQ = PriorityQueue()
    v.path = [1]
    v.bound = bound(v)
    PQ.put((v.bound, v)) #출발노드 push
    printBestFS(v)
    while (not PQ.empty()):
        v = PQ.get()[1]
        if (v.bound < minlength): #pruning
            for i in range(2, n + 1):
                if (v.isIn(i)): #path에 i가 없으면
                        continue
                u = Tree(v.level + 1) #트리 확장
                u.path = deepcopy(v.path)
                u.path.append(i) #추가
                if (u.level < n - 2):
                    printBestFS(u)
                if (u.level == n - 2): #leaf
                    for j in range(2, n + 1):
                        if (not u.isIn(j)):
                            u.path.append(j) #여태 안갔던 노드 추가        
                    u.path.append(1) # 마지막노드(원점) 추가
                    printBestFS(u)
                    if (length(u.path) < minlength):
                        minlength = length(u.path)
                        opttour = deepcopy(u.path) #완성된 path => 최적경로
                            
                else:
                    u.bound = bound(u)
                    if (u.bound < minlength):
                        PQ.put((u.bound, u))
                        
def printBestFS(v): #bestFS 방문노드 출력
    if bound(v) > INF:
        print(v.level,"INF",end=" ")
        for i in range(len(v.path)):
            if i == len(v.path)-1:
                print(v.path[i])
            else:
                print(v.path[i],end=" ")
    else:
        print(v.level,bound(v),end=" ")
        for i in range(len(v.path)):
            if i == len(v.path)-1:
                print(v.path[i])
            else:
                print(v.path[i],end=" ")

#sys.stdin = open("input12-3.txt","r")

n,m = map(int,input().split()) # n = 정점개수 / m = 간선 개수

INF = 999
W = [[INF]*(n+1) for _ in range(n+1)] #인접행렬

for i in range(m):
    u,v,w = map(int,input().split())
    W[u][v]=w
for i in range(n+1):
    W[i][i] = 0

# for i in range(n+1):
#     print(W[i])
    
minlength = INF
opttour = []
v = Tree(0)
travel(v)
#travel(opttour,minlength)
print(minlength)
for i in range(len(opttour)):
    if i == len(opttour)-1:
        print(opttour[i])
    else:
        print(opttour[i],end=" ")
```
